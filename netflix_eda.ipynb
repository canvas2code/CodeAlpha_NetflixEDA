# **Q1: What type of content (Movie or TV Show) is more common on Netflix?**

```python
type_counts = df['type'].value_counts()
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='type', palette='Set2')
plt.title('Content Type Distribution on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

type_counts
```

**Insight:** The chart and count show that Netflix has more Movies than TV Shows.

# **Q2: Which country has produced the most Netflix content?**

```python
top_countries = df['country'].value_counts().dropna().head(10)

top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Most Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
top_countries
```

**Insight:** The US leads by a huge margin, followed by India and the UK  indicating Netflix's strong presence in those markets.(this data is according to 2021 because of no updated data available on website)

# **Q3: What are the top 5 genres on Netflix?**

```python
df['listed_in'] = df['listed_in'].fillna('')
all_genres = df['listed_in'].str.split(', ')
exploded_genres = all_genres.explode()
top_genres = exploded_genres.value_counts().head(5)
top_genres.plot(kind='barh', color='skyblue')
plt.title('Top 5 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
top_genres
```

**Insight:** The US leads by a huge margin, followed by India and the UK. indicating Netflix's strong presence in those markets.

##**Q4: What is the trend of content release over the years on Netflix?**

```python
release_trend = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
release_trend.plot(kind='line', color='green')
plt.title('Netflix Content Release Trend Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
release_trend.tail()
```

**Insight:** Netflix saw a huge content boost between 20152020, peaking before the pandemic hit. This reflects the rise in digital content demand.

# **Q5: Who are the top 10 directors with the most content on Netflix?**

```python
top_directors = df['director'].dropna().value_counts().head(10)
top_directors.plot(kind='barh', color='purple')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
top_directors
```

**Insight:** These directors have multiple releases on Netflix, either through movies or TV shows, showing their consistent collaborations or content licensing.

# **Q6: What is the average duration of Movies vs TV Shows?**

```python
df_movies = df[df['type'] == 'Movie'].copy()
df_movies['duration_num'] = df_movies['duration'].str.extract('(\d+)').astype(float)
df_shows = df[df['type'] == 'TV Show'].copy()
df_shows['seasons_num'] = df_shows['duration'].str.extract('(\d+)').astype(float)
fig, ax = plt.subplots(1, 2, figsize=(12,5))

sns.histplot(df_movies['duration_num'], bins=30, ax=ax[0], color='blue')
ax[0].set_title('Movie Durations (in minutes)')
ax[0].set_xlabel('Duration (minutes)')

sns.histplot(df_shows['seasons_num'], bins=10, ax=ax[1], color='orange')
ax[1].set_title('TV Show Seasons')
ax[1].set_xlabel('Number of Seasons')

plt.tight_layout()
plt.show()
```

**Insight:** Most movies on Netflix are around 90120 minutes long, and TV shows usually run for 12 seasons
.Indicating a preference for short, binge-worthy content.
